```python
def cor_func(bms, col):
    """
    bms is the data frame, col is the target col
    """
    cor1 = pd.DataFrame()
    cor1['ou_code'] = bms['ou_code'].unique()
    list1 = list()
    for ou in bms['ou_code'].unique():
        list1.append(bms[bms['ou_code'] == ou].corr().loc['month',col])
    cor1['key_corr'] = list1
    cor1 = bms.merge(cor1, on = 'ou_code', how = 'left')
    return cor1
```

```python
opera = cor_func(opera, 'lb_op')
store = cor_func(store, 'warehouse_site_fee')
vasty = cor_func(vasty, 'value_added_fee')
balan = cor_func(balan, 'bms_bill_amount_exclude_tax')
```
